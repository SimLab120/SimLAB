```python
import os
import pandas as pd
import numpy as np
from torchvision.io import read_image
from torch.utils.data import Dataset,DataLoader
import torch
# 2 files
# 1259*300*300 ---> 1259*300 in 300 images
# x,y load
# x,y sub sample
# pair
# return
class volDataset(Dataset):
    def __init__(self, x_slices, y_slices, transform=None):
        """
        x_slices: List of noisy seismic slices (each of shape 1259x300)
        y_slices: List of clean seismic slices (each of shape 1259x300)
        """
        self.x_slices = x_slices
        self.y_slices = y_slices
        self.transform = transform

    def __len__(self):
        return len(self.x_slices)

    def __getitem__(self, idx):
        """Returns one slice (1, 1259, 300)"""
        x_slice = torch.tensor(self.x_slices[idx], dtype=torch.float32).unsqueeze(0)  # (1, 1259, 300)
        y_slice = torch.tensor(self.y_slices[idx], dtype=torch.float32).unsqueeze(0)  # (1, 1259, 300)

        if self.transform:
            x_slice = self.transform(x_slice)
            y_slice = self.transform(y_slice)

        return x_slice, y_slice
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models
from torchvision.models import Wide_ResNet101_2_Weights


class Crop2d(nn.Module):
    def __init__(self, left=10, right=10, top=21, bottom=0):
        super(Crop2d, self).__init__()
        self.left = left
        self.right = right
        self.top = top
        self.bottom = bottom

    def forward(self, x):
        return x[:, :, self.top : x.shape[2] - self.bottom, self.left : x.shape[3] - self.right]


class WideResNet101Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        resnet = models.wide_resnet101_2(weights=Wide_ResNet101_2_Weights.IMAGENET1K_V2)
        
        self.initial = nn.Sequential(
            resnet.conv1,
            resnet.bn1,
            resnet.relu,
            resnet.maxpool
        )
        self.layer1 = resnet.layer1  # 1/4 input size
        self.layer2 = resnet.layer2  # 1/8 input size
        self.layer3 = resnet.layer3  # 1/16 input size
        self.layer4 = resnet.layer4  # 1/32 input size
    
    def forward(self, x):
        x = self.initial(x)
        x1 = self.layer1(x)
        x2 = self.layer2(x1)
        x3 = self.layer3(x2)
        x4 = self.layer4(x3)
        
        feature_shapes = [f.shape for f in [x4, x3, x2, x1]]
        
        return [x4, x3, x2, x1]  # Reverse order


class UNetPlusPlusDecoder(nn.Module):
    def __init__(self, num_classes=1):
        super().__init__()
        # Upsampling layers
        self.up1 = nn.ConvTranspose2d(2048, 1024, kernel_size=2, stride=2)  # 40x10 -> 80x20
        self.up2 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)   # 80x20 -> 160x40
        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)    # 160x40 -> 320x80
        self.up4 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)    # 320x80 -> 640x160
        self.up5 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)     # 640x160 -> 1280x320

        # Convolution layers after concatenation
        self.conv1 = nn.Conv2d(1024 + 1024, 1024, kernel_size=3, padding=1)  # up1 + features[1]
        self.conv2 = nn.Conv2d(512 + 512, 512, kernel_size=3, padding=1)     # up2 + features[2]
        self.conv3 = nn.Conv2d(256 + 256, 256, kernel_size=3, padding=1)     # up3 + features[3]
        # For up4 and up5, there are no corresponding skip connections.
        self.conv4 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.conv5 = nn.Conv2d(64, 64, kernel_size=3, padding=1)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def forward(self, features):
        # features[0]: (1, 2048, 40, 10)
        # features[1]: (1, 1024, 80, 20)
        # features[2]: (1, 512, 160, 40)
        # features[3]: (1, 256, 320, 80)
        x = self.up1(features[0])  # → (1, 1024, 80, 20)
        x = torch.cat([x, features[1]], dim=1)  # → (1, 2048, 80, 20)
        x = self.conv1(x)  # → (1, 1024, 80, 20)

        x = self.up2(x)  # → (1, 512, 160, 40)
        x = torch.cat([x, features[2]], dim=1)  # → (1, 1024, 160, 40)
        x = self.conv2(x)  # → (1, 512, 160, 40)

        x = self.up3(x)  # → (1, 256, 320, 80)
        x = torch.cat([x, features[3]], dim=1)  # → (1, 512, 320, 80)
        x = self.conv3(x)  # → (1, 256, 320, 80)

        x = self.up4(x)  # → (1, 128, 640, 160)
        x = self.conv4(x)  # → (1, 128, 640, 160)

        x = self.up5(x)  # → (1, 64, 1280, 320)
        x = self.conv5(x)  # → (1, 64, 1280, 320)

        x = self.final_conv(x)  # → (1, 1, 1280, 320)
        return x


class CNNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.zero_pad = nn.ZeroPad2d((10, 10, 21, 0))
        self.crop = Crop2d(left=10, right=10, top=21, bottom=0)
        self.encoder = WideResNet101Encoder()
        self.decoder = UNetPlusPlusDecoder()

    def forward(self, x):
        x = self.zero_pad(x)          # Input (1,1,1259,300) becomes (1,1,1280,320)
        x = x.expand(-1, 3, -1, -1)     # Expand to 3 channels → (1,3,1280,320)
        features = self.encoder(x)
       
        x = self.decoder(features)      # Should become (1,1,1280,320)
        x = self.crop(x)                # Crop to remove zero pad → (1,1,1259,300)
       
        return x


# Test the model
model = CNNModel()
x = torch.randn(1, 1, 1259, 300)  # Example input (1,1,1259,300)
out = model(x)
if out is not None:
    print(out.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define model
model = CNNModel()

# Define Loss Function (MSE for denoising)
criterion = nn.MSELoss()

# Define Optimizer (Adam works well)
optimizer = optim.Adam(model.parameters(), lr=0.1)
```

```python
# Check batch shapes
for x_batch, y_batch in train_loader:
    print("Train Batch Shape:", x_batch.shape, y_batch.shape)  # Expected: (batch_size, 1, 1259, 300)
    break
```

```python
from torchmetrics import StructuralSimilarityIndexMeasure
ssim_metric = StructuralSimilarityIndexMeasure()
```

```python
import matplotlib.pyplot as plt
```

```python
# Number of epochs
num_epochs = 200

# Training loop
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_ssim = 0.0

    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(x_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        batch_ssim = ssim_metric(outputs.detach(), y_batch.detach())
        running_ssim += batch_ssim.item()

    # Validation phase
    model.eval()
    val_loss = 0.0
    val_ssim = 0.0
    with torch.no_grad():
        for x_val, y_val in val_loader:
            val_outputs = model(x_val)
            val_loss += criterion(val_outputs, y_val).item()
            val_ssim += ssim_metric(val_outputs, y_val).item()

    # Prediction on a single slice from X_slices_t
    random_slice = X_slices_t[torch.randint(0, len(X_slices_t), (1,)).item()]

    # Convert to PyTorch tensor and add batch and channel dimensions
    random_slice = torch.tensor(random_slice, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

    prediction = model(random_slice).detach()

    # Visualize the original and predicted slice
    plt.subplot(1, 2, 1)
    plt.title("Original Slice")
    plt.imshow(random_slice.squeeze().cpu().numpy(), cmap='seismic')

    plt.subplot(1, 2, 2)
    plt.title("Predicted Slice")
    plt.imshow(prediction.squeeze().cpu().numpy(), cmap='seismic')
    plt.show()

    print(f"Epoch [{epoch + 1}/{num_epochs}], Train Loss: {running_loss / len(train_loader):.4f}, Train SSIM: {running_ssim / len(train_loader):.4f}, \
          Val Loss: {val_loss / len(val_loader):.4f}, Val SSIM: {val_ssim / len(val_loader):.4f}")


print("Training complete!")
```

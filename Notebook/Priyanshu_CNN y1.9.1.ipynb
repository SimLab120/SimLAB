```python
import os
import pandas as pd
import numpy as np
from torchvision.io import read_image
from torch.utils.data import Dataset,DataLoader
import torch
# 2 files
# 1259*300*300 ---> 1259*300 in 300 images
# x,y load
# x,y sub sample
# pair
# return
class volDataset(Dataset):
    def __init__(self, x_slices, y_slices, transform=None):
        """
        x_slices: List of noisy seismic slices (each of shape 1259x300)
        y_slices: List of clean seismic slices (each of shape 1259x300)
        """
        self.x_slices = x_slices
        self.y_slices = y_slices
        self.transform = transform

    def __len__(self):
        return len(self.x_slices)

    def __getitem__(self, idx):
        """Returns one slice (1, 1259, 300)"""
        x_slice = torch.tensor(self.x_slices[idx], dtype=torch.float32).unsqueeze(0)  # (1, 1259, 300)
        y_slice = torch.tensor(self.y_slices[idx], dtype=torch.float32).unsqueeze(0)  # (1, 1259, 300)

        if self.transform:
            x_slice = self.transform(x_slice)
            y_slice = self.transform(y_slice)

        return x_slice, y_slice
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models
from torchvision.models import VGG11_BN_Weights


class Crop2d(nn.Module):
    def __init__(self, left=10, right=10, top=21, bottom=0):
        super(Crop2d, self).__init__()
        self.left = left
        self.right = right
        self.top = top
        self.bottom = bottom

    def forward(self, x):
        return x[:, :, self.top:x.shape[2] - self.bottom, self.left:x.shape[3] - self.right]


class VGG16BNEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        vgg = models.vgg11_bn(weights=VGG11_BN_Weights.IMAGENET1K_V1)

        self.encoder = vgg.features
        self.encoder[0] = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1, bias=False)
        self.encoder[0].weight.data = vgg.features[0].weight.data.mean(dim=1, keepdim=True)

        self.layer1 = self.encoder[:4]   # out: 64
        self.layer2 = self.encoder[4:8]  # out: 128
        self.layer3 = self.encoder[8:12] # out: 256
        self.layer4 = self.encoder[12:16]# out: 512
        self.layer5 = self.encoder[16:]  # out: 512


    def forward(self, x):
        x1 = self.layer1(x)  # 1/2 input size
        x2 = self.layer2(x1) # 1/4 input size
        x3 = self.layer3(x2) # 1/8 input size
        x4 = self.layer4(x3) # 1/16 input size
        x5 = self.layer5(x4) # 1/32 input size

        


        return [x5, x4, x3, x2, x1]  # Reverse order


class UNetPlusPlusDecoder(nn.Module):
    def __init__(self, num_classes=1):
        super().__init__()

        self.up1 = nn.ConvTranspose2d(512, 512, kernel_size=2, stride=2)  # x5 → x4 (both 512)
        self.up2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)  # x4 → x3 (target 256)
        self.up3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)  # x3 → x2 (target 128)
        self.up4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)   # x2 → x1 (target 64)

        self.final_conv = nn.Conv2d(64, num_classes, kernel_size=1)

    def forward(self, features):
        x = F.interpolate(self.up1(features[0]), size=features[1].shape[2:], mode="bilinear", align_corners=False) + features[1]
        x = F.interpolate(self.up2(x), size=features[2].shape[2:], mode="bilinear", align_corners=False) + features[2]
        x = F.interpolate(self.up3(x), size=features[3].shape[2:], mode="bilinear", align_corners=False) + features[3]
        x = F.interpolate(self.up4(x), size=features[4].shape[2:], mode="bilinear", align_corners=False) + features[4]
        x = self.final_conv(x)

        # NEW: Final upsampling to match original padded input size
        x = F.interpolate(x, scale_factor=2, mode="bilinear", align_corners=False)

        return x




class CNNModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.zero_pad = nn.ZeroPad2d((10, 10, 21, 0))
        self.crop = Crop2d(left=10, right=10, top=21, bottom=0)

        self.encoder = VGG16BNEncoder()
        self.decoder = UNetPlusPlusDecoder()

    def forward(self, x):
        x = self.zero_pad(x)

        # Get feature maps from encoder
        features = self.encoder(x)

        
        
        # Decode
        x = self.decoder(features)

        # Apply crop to remove zero padding
        x = self.crop(x)
        return x


# Test the model
model = CNNModel()
x = torch.randn(1, 1, 1259, 300)  # Example grayscale input
out = model(x)
if out is not None:
    print(out.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define model
model = CNNModel()

# Define Loss Function (MSE for denoising)
criterion = nn.MSELoss()

# Define Optimizer (Adam works well)
optimizer = optim.Adam(model.parameters(), lr=0.1)
```

```python
# Check batch shapes
for x_batch, y_batch in train_loader:
    print("Train Batch Shape:", x_batch.shape, y_batch.shape)  # Expected: (batch_size, 1, 1259, 300)
    break
```

```python
from torchmetrics import StructuralSimilarityIndexMeasure
ssim_metric = StructuralSimilarityIndexMeasure()
```

```python
import matplotlib.pyplot as plt
```

```python
# Number of epochs
num_epochs = 200

# Training loop
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_ssim = 0.0

    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(x_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        batch_ssim = ssim_metric(outputs.detach(), y_batch.detach())
        running_ssim += batch_ssim.item()

    # Validation phase
    model.eval()
    val_loss = 0.0
    val_ssim = 0.0
    with torch.no_grad():
        for x_val, y_val in val_loader:
            val_outputs = model(x_val)
            val_loss += criterion(val_outputs, y_val).item()
            val_ssim += ssim_metric(val_outputs, y_val).item()

    # Prediction on a single slice from X_slices_t
    random_slice = X_slices_t[torch.randint(0, len(X_slices_t), (1,)).item()]

    # Convert to PyTorch tensor and add batch and channel dimensions
    random_slice = torch.tensor(random_slice, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

    prediction = model(random_slice).detach()

    # Visualize the original and predicted slice
    plt.subplot(1, 2, 1)
    plt.title("Original Slice")
    plt.imshow(random_slice.squeeze().cpu().numpy(), cmap='seismic')

    plt.subplot(1, 2, 2)
    plt.title("Predicted Slice")
    plt.imshow(prediction.squeeze().cpu().numpy(), cmap='seismic')
    plt.show()

    print(f"Epoch [{epoch + 1}/{num_epochs}], Train Loss: {running_loss / len(train_loader):.4f}, Train SSIM: {running_ssim / len(train_loader):.4f}, \
          Val Loss: {val_loss / len(val_loader):.4f}, Val SSIM: {val_ssim / len(val_loader):.4f}")


print("Training complete!")
```
